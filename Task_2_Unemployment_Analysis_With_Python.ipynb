## Analyzing Data using heatmap

```python
plt.figure(figsize=(12, 8))
sns.heatmap(data[[' Estimated Unemployment Rate (%)',
 ' Estimated Employed',
 ' Estimated Labour Participation Rate (%)','longitude',
 'latitude']].corr())
plt.title("Correlation Heatmap")
plt.show()
```

## Analyzing Data by bar graphs

```python
graph_rep = pex.bar(data,x='Region',y=' Estimated Unemployment Rate (%)',color='Region',
          title='Unemployment Rate (State-Wise) by bar Graph',template='plotly')
graph_rep.update_layout(xaxis={'categoryorder':'total descending'})
graph_rep.show()
```

```python
graph_rep = pex.bar(data,x='Region.1',y=' Estimated Unemployment Rate (%)',color='Region',
          title='Unemployment Rate (Region-Wise) by Bar Graph',template='plotly')
graph_rep.update_layout(xaxis={'categoryorder':'total descending'})
graph_rep.show()
```

```python
graph_rep = pex.box(data,x='Region',y=' Estimated Unemployment Rate (%)',color='Region',
          title='Unemployment Rate (State-Wise) by box Graph',template='plotly')
graph_rep.update_layout(xaxis={'categoryorder':'total descending'})
graph_rep.show()
```

```python
graph_rep = pex.scatter(data,x='Region',y=' Estimated Unemployment Rate (%)',color='Region',
          title='Unemployment Rate (State-Wise) by Scatter Graph',template='plotly')
graph_rep.update_layout(xaxis={'categoryorder':'total descending'})
graph_rep.show()
```

```python
graph_rep = pex.histogram(data,x='Region',y=' Estimated Unemployment Rate (%)',color='Region',
          title='Unemployment Rate (State-Wise) by Histogram',template='plotly')
graph_rep.update_layout(xaxis={'categoryorder':'total descending'})
graph_rep.show()
```
